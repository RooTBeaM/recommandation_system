```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_product = pd.read_csv('./csv/dtt_product.csv')
print(f'Total rows : {len(df_product)}')
df_product.sample(3)
```

```python
# Drop NaN columns
drop_cols = df_product.columns[df_product.isnull().sum()>100].tolist()
df_product.drop(columns = drop_cols, inplace=True)

# selected columns
product_cols = [
       'product_id', 'product_slug','age_adult', 'age_child', 'age_infant',
       'is_promotion', 'is_recommend', 'is_not_available', 
       'subdistrict_id', 'province_id', 'district_id', 'location_id', 
       'product_lat', 'product_long', 
       'browser', 'platform',
       'country_id', 'view_count'
       ]
df_product = df_product[product_cols]
df_product.sample(3)
```
