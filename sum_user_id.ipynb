```python
def softmax(x, axis = 1):
    return np.exp(x)/np.sum(np.exp(x),axis=axis, keepdims=True)
```

```python
use_cols = [
    'new_id',
    'booked_days', 'order_price_paid', 'sum_kids', 'sum_adults', # mean
    'private', 'group', 'family', 
    'cat_1', 'cat_2', 'cat_3', 'cat_4', 'cat_5', 'cat_6', 'cat_7', 'cat_8', 'cat_9',
    'cat_10', 'cat_11', 'cat_12', 'cat_13', 'cat_14', 'cat_15', 'cat_16','cat_17', 'cat_18'
    ]
cat_cols = ['cat_1', 'cat_2', 'cat_3', 'cat_4', 'cat_5', 'cat_6', 'cat_7', 'cat_8', 'cat_9',
    'cat_10', 'cat_11', 'cat_12', 'cat_13', 'cat_14', 'cat_15', 'cat_16','cat_17', 'cat_18']
```

```python
df_matrix_user = df_clean_userID[use_cols]
df_matrix_user[cat_cols] = softmax(df_matrix_user[cat_cols].applymap(lambda x: 0.3 if x == 0 else x).to_numpy())
df_matrix_user[['private', 'group', 'family']] = softmax(df_matrix_user[['private', 'group', 'family']].to_numpy())
df_matrix_user
```
